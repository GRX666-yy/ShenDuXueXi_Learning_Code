```python
%matplotlib inline
import torch
import torchvision
from torch import nn
from d2l import torch as d2l
```

```python
d2l.set_figsize()
img = d2l.Image.open("D:\\Pytorch\\data\\cat_and_dog\\train\\cats\\cat.59.jpg")
d2l.plt.imshow(img)
```

```python
def apply(img,aug,num_rows=2,num_cols=4,scale=1.5):
    Y=[aug(img) for _ in range(num_rows*num_cols)]
    d2l.show_images(Y,num_rows,num_cols,scale=scale)
```

```python
#水平翻转
H_flip_aug=apply(img,torchvision.transforms.RandomHorizontalFlip())
```

```python
#上下翻转，注意图片是否适合上下翻转
V_flip_aug=apply(img,torchvision.transforms.RandomVerticalFlip())
```

```python
#随机裁剪
clip_aug=torchvision.transforms.RandomResizedCrop((200,200),scale=(0.1,1),ratio=(0.5,2))
apply(img,clip_aug)
```

```python
#随机亮度
color_aug=torchvision.transforms.ColorJitter(brightness=0,contrast=0,saturation=0,hue=0.5)
apply(img,color_aug)
```

```python
augs=torchvision.transforms.Compose([torchvision.transforms.RandomHorizontalFlip(),clip_aug,color_aug])
apply(img,augs)
```
